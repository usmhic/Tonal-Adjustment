```python
from skimage.io import *
import sys
import photomosaic
from skimage import data
import matplotlib.pyplot as plt
import cv2

image = cv2.imread('../assets/image-samples/img2.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image) 

photomosaic.rainbow_of_squares('../assets/image-samples/img2.jpg')
square_pool = photomosaic.make_pool('../assets/image-samples/img2.jpg') 

mosaic = photomosaic.basic_mosaic(image, square_pool,(200, 300))
plt.imshow(mosaic)
```
